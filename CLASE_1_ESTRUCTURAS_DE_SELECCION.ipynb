##1. IF/ELSE
Sólo tiene dos posibles respuestas (V) o (F). Sintaxis


```
if (condicion):
  aquí se ejecuta código **si** la condicion es verdadera.
else:
  aquí se ejecuta el código **si** la condicion es falsa.
```
