#ESTRUCTURA WHILE - MIENTRAS
No se sabe cuantas veces se debe repetir la acción

Sintaxis en phyton
```
while (CI):
  parte del codigo a repetir
  Modificador de CI
```
**OJO!!**
CI: La CI debe tener un valor de verdad VERDADERO(TRUE) para poder ingresar al while
CI=Condición Inicial



Sintaxis en java

```
while(CI){
  Código que debe repetirse
  Un modificador de CI
}
```



````python
#Piedra, papel o tijera
import random
yo=int(input("Elige entre 1.Piedra, 2.Papel, 3.Tijera"))

````

````python
m=random.randint(1,3)
m
````

````python
if yo==m: print("Empate. Ambos elegimos 🪨")
elif yo==1 and m==2 :print("Perdiste. Maquina eligió📋")
elif yo==2 and m==3: print("Perdiste. Maquina eigió📋")
else: print("Ganaste!!")
````
