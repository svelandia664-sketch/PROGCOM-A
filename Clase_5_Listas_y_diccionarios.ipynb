`````python
#De una lista n trabajadores, decir a. Cuánto ganan más que el promedio y quienes son.
n=int(input("Ingrese el número de trabajadores: "))

nombres=[]
sueldos=[]

for i in range(n):
    nombre = input(f"Ingrese el nombre del trabajador {i+1} :")
    sueldo = float(input(f"Ingrese el sueldo del trabajador {i+1}:" ))

    nombres.append(nombre)
    sueldos.append(sueldo)

suma=sum(sueldos)
prom=suma/n

c=0

print("Trabajadores que ganan más que el promedio:")

for i in range(n):
    if sueldos[i] > prom:
        print(nombres[i])
        c=c + 1

print("Cantidad que ganan más que el promedio:", c)
`````

`````python
#Desarrollar el menu de una tienda empleando listas
n=int(input("Ingrese el número de productos: "))
productos=[]
precios=[]
for i in range(n):
    producto=input(f"Ingrese el nombre del producto {i+1}: ")
    precio=float(input(f"Ingrese el precio del producto {i+1}: "))

    productos.append(producto)
    precios.append(precio)

print("Lista de productos disponibles:")

for i in range(n):
    print(productos[i], ":", precios[i])

c=0
total=0

m=int(input("Ingrese la cantidad de productos que desea comprar: "))

for i in range(m):
    compra=input(f"Ingrese el nombre del producto a comprar {i+1}: ")

    for j in range(n):
        if compra == productos[j]:
            print("Producto encontrado - Precio:", precios[j])
            total = total + precios[j]
            c = c + 1

print("Cantidad de productos comprados:", c)
print("Total a pagar:", total)
`````

`````python
#Decir cuántos elementos de una lista son múltiplos del números n
numeros = []

n_elementos = int(input("¿Cuántos números desea ingresar? "))

print("\n=== INGRESO DE NÚMEROS ===\n")

for i in range(n_elementos):
    num = int(input(f"Ingrese el número {i+1}: "))
    numeros.append(num)

n = int(input("\n¿De qué número desea encontrar los múltiplos? "))

multiplos = [x for x in numeros if x % n == 0]
contador = len(multiplos)

print("\n=== RESULTADO ===\n")

for num in numeros:
    if num % n == 0:
        print(f"✔ {num} - Es múltiplo de {n}")
    else:
        print(f"✘ {num} - No es múltiplo de {n}")

print(f"\nTotal de números ingresados: {len(numeros)}")
print(f"Múltiplos de {n}:            {contador}")
print(f"No múltiplos de {n}:         {len(numeros) - contador}")
`````

`````python
#Indicar cuántos de n pacientes han sido vacunados contra el covid. Mencionar fechas de vacunación
pacientes = []
vacunados = []
fechas = []

n = int(input("¿Cuántos pacientes desea registrar? "))

print("\n=== INGRESO DE PACIENTES ===\n")

for i in range(n):
    print(f"Paciente {i+1} ")
    nombre = input("Nombre del paciente: ")
    estado = input("¿Está vacunado? (si/no): ").lower()

    pacientes.append(nombre)

    if estado == "si":
        fecha = input("Fecha de vacunación (DD/MM/AAAA): ")
        vacunados.append(True)
        fechas.append(fecha)
    else:
        vacunados.append(False)
        fechas.append(None)

    print()

contador = 0

print("=== REGISTRO DE VACUNACIÓN COVID ===\n")

for i in range(len(pacientes)):
    if vacunados[i] == True:
        contador += 1
        print(f"✔ {pacientes[i]} - Vacunado el: {fechas[i]}")
    else:
        print(f"✘ {pacientes[i]} - No vacunado")

print(f"\nTotal de pacientes:   {len(pacientes)}")
print(f"Pacientes vacunados:  {contador}")
print(f"Pacientes sin vacunar:{len(pacientes) - contador}")
`````

`````python
#Ejercicio 2
candidatos = ["Ana", "Carlos", "María", "Pedro", "Lucía"]
def votar(lista):
    votos = [0] * len(lista)
    for i in range(int(input("¿Cuántos votan? "))):
        print("\nCandidatos:")
        for j in range(len(lista)):
            print(f"  {j+1}. {lista[j]} ({votos[j]} votos)")
        votos[int(input("Voto: ")) - 1] += 1
    return votos
v1 = votar(candidatos)
print("RESULTADOS PRIMERA VUELTA")
for i in range(len(candidatos)):
    print(f"  {candidatos[i]}: {v1[i]} votos")
max_votos = max(v1)
finalistas = []
for i in range(len(candidatos)):
    if v1[i] == max_votos:
        finalistas.append(candidatos[i])
print(f"\nFinalistas: {', '.join(finalistas)}")
v2 = votar(finalistas)
print("RESULTADOS SEGUNDA VUELTA")
for i in range(len(finalistas)):
    print(f"  {finalistas[i]}: {v2[i]} votos")
if v2.count(max(v2)) > 1:
    print(f"\n🤝 EMPATE")
else:
    ganador = finalistas[v2.index(max(v2))]
    print(f"\n🏆 Ganador: {ganador}")
`````
