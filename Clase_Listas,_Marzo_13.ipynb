##FUNCIÓN
Es un pedazo de código que se ejecuta aparte. A veces se conoce como copiado dinámico.

Ventajas de implementar las funciones
1. Vuelve el código más limpio
2. Elimina el código espaguetti
3. Establece un orden de búsqueda
4. Permite reutilización del código

Sintaxis
```
def nombre_de_la_función(parámetros):
  Aqui va todo lo que desees ejecutar
  return algo
```
Existen 4 tipo de funciones
1. Sin parámetros, sin retorno
2. Sin parámetros, con retorno
3. Con parámetros, sin retorno
4. Con parámetros, con retorno
