##DICCIONARIO
1. Tipo de dato como las lsitas. **SI SON MUTABLES**
2. Se utilizan para almacenar datos en pares,
clave:valor. La clave **DEBE SER ALFANUMÉRICA**
```
variable={clave:valor}
```
3. Los diccionarios son colecciones ordenadas, modificables y **NO PERMITE DUPLICADOS**.


````python
estudiante = {"nombre": "simon","ID": "U00186384", "Edad": "18","Carrera": "Ingenieria Industrial"
}

print(estudiante)
````

````python
galletas=["angelitos","diablitos","dinosaurios"]
galletas_precio=[20000,15000,25000]
galletas_cant=[3,5,10]
````

````python
def menu_galletas():
  for i in range(3):
    print(f"{galletas[i]}:{galletas_precio[i]}, cantidad:{galletas_cant[i]}")
````

````python
menu_galletas()
````

````python
galleta={"tipo de galleta":["angelitos","diablitos","dinosaurios"],
         "precio":[20000,15000,25000],
         "cantidad":[3,5,10]}
````

```

```
2. Quiero imprimir las claves
```
print(diccionario.keys)
```

````python
def menu_galletas_dict():
  for i in range(3):
    print(f"{galleta["tipo de galleta"][i]}:{galleta["precio"][i]}, cantidad:{galleta["cantidad"][i]}")
````

````python
menu_galletas_dict()
````
