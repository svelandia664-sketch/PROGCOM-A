```python
#Ejercicio 22
numero = int(input("Ingrese un número: "))
suma = 0

while numero > 0:
    digito = numero % 10
    suma = suma + digito
    numero = numero // 10

print("La suma de los dígitos es:", suma)
```

```python
#Ejercicio 23
num = int(input("Ingrese un número: "))
i = 1
contador = 0

while i <= num:
    if num % i == 0:
        contador = contador + 1
    i = i + 1

if contador == 2:
    print("Es primo")
else:
    print("No es primo")
```

```python
#Ejercicio 24
filas = int(input("Ingrese el número de filas: "))

i = 1

while i <= filas:
    linea = ""
    j = 1

    while j <= i:
        linea = linea + "*"
        j = j + 1

    print(linea)
    i = i + 1
```
