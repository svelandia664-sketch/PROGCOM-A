```python
#Ejercicio 23
n = int(input("Ingrese un número: "))
contador = 0

for i in range(1, n + 1):
    if n % i == 0:
        contador = contador + 1

if contador == 2:
    print("Es primo")
else:
    print("No es primo")
```
