##1. para - for

* Se emplea cuando se sabe cuantas veces se repetirá la acción.
* **OJO** El conteo siempre inicia desde **0**.

Sintaxis general:
```
# for (variable) in (algo):
  La acción que se va a repetir

#Forma2
for variable in range(cantidad de veces):
  La acción que se va a repetir.

#Forma3
for variable in range(VI,VF+1,P):
  La acción que se va a repetir.
```
Donde:
* VI: Valor inicial (dónde inicio).
* VF: Valor final (dónde finalizo).
* P: Paso. Cáda cuánto cuento



Sintaxis Para en java

```
for (int variable=VI;varaible=VF;variable=variable+paso){
  La accion que se va a repetir.}
```



````python
#Ejercicio: Imprimir los números pares del 1 al 10
for num in range(11):
  if num%2==0: print(num)
````

````python
for num in range (0,11):
  if num%2==0: print(num)
````

````python
for num in range(0,11,2):
  print(num)
````
