```python
tareas = []

while True:
    print("\n--- MENÚ ---")
    print("1. Agregar tarea")
    print("2. Ver tareas")
    print("3. Salir")

    opcion = input("Elige una opción: ")

    if opcion == "1":
        tarea = input("Escribe la tarea: ")
        tareas.append(tarea)

    elif opcion == "2":
        if len(tareas) == 0:
            print("No hay tareas.")
        else:
            for i, t in enumerate(tareas):
                print(f"{i+1}. {t}")

    elif opcion == "3":
        print("Adiós 👋")
        break

    else:
        print("Opción inválida")
```
