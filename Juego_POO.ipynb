```python
"""
Juego educativo: Aventura POO
Autor: Luisa Fernanda Toro Fuentes
Año: 2026
"""

import random

# COLORES
verde = "\033[92m"
rojo = "\033[91m"
azul = "\033[94m"
amarillo = "\033[93m"
reset = "\033[0m"


# -------------------------
# CLASE JUGADOR
# -------------------------

class Jugador:

    def __init__(self, nombre):
        self.nombre = nombre
        self.vidas = 3
        self.puntos = 0

    def mostrar_vidas(self):
        print("Vidas:", "❤️" * self.vidas)


# -------------------------
# CLASE PREGUNTA
# -------------------------

class Pregunta:

    def __init__(self, texto, respuesta, explicacion):
        self.texto = texto
        self.respuesta = respuesta
        self.explicacion = explicacion


# -------------------------
# CLASE NIVEL
# -------------------------

class Nivel:

    def __init__(self, nombre, preguntas):
        self.nombre = nombre
        self.preguntas = preguntas


# -------------------------
# CLASE JUEGO
# -------------------------

class Juego:

    def __init__(self, jugador, niveles):
        self.jugador = jugador
        self.niveles = niveles


    # MINIJUEGO PARA RECUPERAR VIDAS
    def minijuego(self):

        print(rojo + "\nTe quedaste sin vidas" + reset)
        print("Completa correctamente para recuperar vidas\n")

        retos = [
            ("Una ______ es una plantilla para crear objetos", "clase"),
            ("Una instancia de una clase es un ______", "objeto"),
            ("La ______ permite reutilizar código entre clases", "herencia"),
            ("El ______ protege los datos de una clase", "encapsulamiento"),
        ]

        retos_disponibles = retos.copy()
        random.shuffle(retos_disponibles)

        while self.jugador.vidas < 3:

            if len(retos_disponibles) == 0:
                retos_disponibles = retos.copy()
                random.shuffle(retos_disponibles)

            reto = retos_disponibles.pop()

            respuesta = input(reto[0] + " : ")

            if respuesta.lower() == reto[1]:

                print(verde + "Correcto! +1 vida" + reset)
                self.jugador.vidas += 1

            else:

                print(rojo + "Incorrecto intenta otra vez" + reset)

            print("Vidas actuales:", self.jugador.vidas)


    # JUEGO PRINCIPAL
    def jugar(self):

        for nivel in self.niveles:

            nivel_superado = False

            while not nivel_superado:

                print("\n=====================")
                print("NIVEL: " + nivel.nombre)
                print("=====================")

                correctas = 0

                for pregunta in nivel.preguntas:

                    self.jugador.mostrar_vidas()

                    print("\n" + pregunta.texto)

                    respuesta = input("Respuesta: ")

                    if respuesta.lower() == pregunta.respuesta.lower():

                        print(verde + "Correcto" + reset)
                        self.jugador.puntos += 10
                        correctas += 1

                    else:

                        print(rojo + "Incorrecto" + reset)
                        print("Explicación:", pregunta.explicacion)
                        self.jugador.vidas -= 1

                    if self.jugador.vidas == 0:
                        self.minijuego()

                if correctas >= len(nivel.preguntas) // 2:

                    print(verde + "\nPasaste el nivel!" + reset)
                    nivel_superado = True

                else:

                    print(azul + "\nNo respondiste todo o la mayoria bien" + reset)
                    print(azul + "Debes repetir este nivel.\n" + reset)

        print("\nJuego terminado")
        print("Puntaje final:", self.jugador.puntos)


# -------------------------
# PREGUNTAS
# -------------------------

nivel1 = Nivel("Clases y Objetos", [

    Pregunta(
        "¿Cómo se llama la plantilla para crear objetos?",
        "clase",
        "Una clase es una estructura que define los datos y métodos que tendra un objeto."
    ),

    Pregunta(
        "Una instancia de una clase es un:",
        "objeto",
        "Un objeto es una instancia concreta de una clase, representando una entidad específica."
    ),

    Pregunta(
        "¿Las clases permiten organizar código? (si/no)",
        "si",
        "Las clases ayudan a estructurar mejor los programas."
    )
])


nivel2 = Nivel("Herencia", [

    Pregunta(
        "¿Qué concepto permite que una clase herede características de otra?",
        "herencia",
        "La herencia permite reutilizar código entre clases."
    ),

    Pregunta(
        "¿La herencia ayuda a reutilizar código? (si/no)",
        "si",
        "Gracias a la herencia podemos evitar repetir código."
    ),

    Pregunta(
        "La clase que hereda se llama clase:",
        "hija",
        "La clase hija recibe propiedades de la clase padre."
    )
])


nivel3 = Nivel("Encapsulamiento y Polimorfismo", [

    Pregunta(
        "¿Qué protege los datos dentro de una clase?",
        "encapsulamiento",
        "El encapsulamiento controla el acceso a los datos."
    ),

    Pregunta(
        "¿Qué permite que un método tenga diferentes comportamientos?",
        "polimorfismo",
        "El polimorfismo permite que un mismo método actúe diferente."
    ),

    Pregunta(
        "¿El polimorfismo mejora la flexibilidad del código? (si/no)",
        "si",
        "Permite reutilizar métodos con comportamientos diferentes."
    )
])


# -------------------------
# INICIO DEL JUEGO
# -------------------------

print(amarillo + "Bienvenido a Aventura POO" + reset)

nombre = input("Ingresa tu nombre: ")

jugador = Jugador(nombre)

niveles = [nivel1, nivel2, nivel3]

juego = Juego(jugador, niveles)

juego.jugar()
```
