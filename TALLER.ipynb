```python
"EJERCICIO 5"
estudiantes = {
    "Ana": 4.5,
    "Luis": 3.8,
    "Sofia": 4.2
}

def promedio_general(dic):
    suma = sum(dic.values())
    return suma / len(dic)

def promedio_estudiante(dic, nombre):
    return dic[nombre]

print("Promedio general:", promedio_general(estudiantes))
print("Nota de Ana:", promedio_estudiante(estudiantes, "Ana"))
```

```python
"EJERCICIO 7"
palabra = "magia"
conteo = {}

for letra in palabra:
    if letra in conteo:
        conteo[letra] += 1
    else:
        conteo[letra] = 1

print(conteo)
```

```python
"EJERCICIO 8"
lista = ["gema", "oro", "gema", "madera", "oro", "oro", "gema"]
conteo = {}

for item in lista:
    if item in conteo:
        conteo[item] += 1
    else:
        conteo[item] = 1

print(conteo)
```
