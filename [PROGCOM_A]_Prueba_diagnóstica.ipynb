```python
#Punto 15 - if else
a = int(input("Lado 1: "))
b = int(input("Lado 2: "))
c = int(input("Lado 3: "))

if a + b > c and a + c > b and b + c > a:
    if a == b and b == c:
        print("Sí forman triángulo - Equilátero")
    else:
        if a == b or a == c or b == c:
            print("Sí forman triángulo - Isósceles")
        else:
            print("Sí forman triángulo - Escaleno")
else:
    print("No forman triángulo")
```
