#DICCIONARIOS
1. Tipo de dato
2. **NO TIENEN INDEX**. Tienen clave:valor
3. Las claves **SIEMPRE** son caracteres.
4. Los valores pueden ser cualquier otro tipo de dato

Sintaxis
```
variable={"clave1":valor,"clave2":valor,...,"claven":valor}
```



````python
galleta=["dinosaurio","flores","carro"]
precio=[5000,6000,7000]
cant=[5,5,6]
````

````python
def menu():
  for galletas in range(3):
    print(f"{galleta[galletas]}, precio={precio[galletas]}, cantidad={cant[galletas]}")
````

````python
menu()
````
